```python
import re
from collections import Counter
```

```python
print("TASK 1: CONCEPTUAL UNDERSTANDING\n")

print("1. Difference between 'Love' and 'love':")
print("In NLP, they are treated as different tokens due to case sensitivity. After preprocessing (lowercasing), both become 'love'.\n")

print("2. What happens if stopwords are not removed?")
print("Text becomes noisy, increases size, and reduces model efficiency as common words dominate.\n")

print("3. Cases where removing stopwords is harmful:")
print("- Sentiment Analysis: 'not' changes meaning")
print("- Question Answering: removes intent words like 'what', 'why'\n")

print("4. Stemming vs Lemmatization:")
print("Stemming: cuts words (running -> run)")
print("Lemmatization: returns meaningful base form (better -> good)\n")
```

```python
# TASK 2: PREPROCESS FUNCTION

print("\n===== TASK 2: PREPROCESS FUNCTION TEST =====\n")

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Lowercase
    text = text.lower()

    # Remove special characters
    text = re.sub(r"[^\w\s]", "", text)

    # Remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    # Remove short tokens ≤2 except 'no', 'not'
    tokens = [w for w in tokens if len(w) > 2 or w in ['no', 'not']]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence


# TEST CASES FOR TASK 2
test_cases_task2 = [
    "I have 2 dogs",
    "This   is   good",
    "soooo goooood!!!",
    "WOW!!! This is GREAT!!!",
    "Visit http://example.com now",
]

for t in test_cases_task2:
    tokens, clean = preprocess_text(t)
    print("Input :", t)
    print("Output:", clean)
    print("-"*50)
```

```python
# TASK 3: STRESS TESTING

print("TASK 3: STRESS TESTING")

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)

    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", clean)
    print("-"*60)

    results.append((sentence, tokens, clean))
```

```python
# TASK 4: TOKEN ANALYTICS

print("TASK 4: TOKEN ANALYTICS")

def token_analytics(tokens):
    if not tokens:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(w) for w in tokens) / total

    return total, unique, round(avg_len, 2)


for sentence, tokens, _ in results:
    total, unique, avg = token_analytics(tokens)

    print("Sentence:", sentence)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Avg Token Length:", avg)
    print("-"*60)


print("\nAnalysis:")
print("Most noisy sentence: Win $$$ now!!! Limited offer!!!")
print("Most meaningful sentence: I am not happy with this")
```

```python
# TASK 5: FREQUENCY ANALYSIS

print("TASK 5: FREQUENCY ANALYSIS")

all_tokens = []
for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
# TASK 6: FULL PIPELINE

print("TASK 6: FULL PIPELINE")

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

pipeline_output = full_pipeline(test_sentences)
print(pipeline_output)


# TEST CASE FOR PIPELINE
print("\nPipeline Test Case:")
sample = ["Hello!!! 123 world", "Nooooo wayyyy!!!"]
print(full_pipeline(sample))
```

```python
# TASK 7: ERROR HANDLING

print("TASK 7: ERROR HANDLING")

edge_cases = ["", "😍😍😍", "123456789"]

for case in edge_cases:
    tokens, clean = preprocess_text(case)
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-"*50)


print("\n✅ ALL TASKS EXECUTED SUCCESSFULLY")
```
