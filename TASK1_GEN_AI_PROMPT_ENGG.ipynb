```python
# ============================
# Install
# ============================
!pip install -q langchain langchain-core langchain-community requests==2.32.5

# ============================
# Imports
# ============================

from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

# ============================
# Task 1: Replace Hardcoded Prompts
# ============================

template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print("Task 1 Output:")
print(template_1.format(topic="Artificial Intelligence"))
print("\n")


# ============================
# Task 2: Multi-Input Prompt System
# ============================

multi_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print("Task 2 Outputs:")
print(multi_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
print("\n")


# ============================
# Task 3: Prompt Variations Engine
# ============================

teach_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic_val = "Machine Learning"

print("Task 3 Outputs:")
print(teach_template.format(topic=topic_val))
print(interview_template.format(topic=topic_val))
print(story_template.format(topic=topic_val))
print("\n")


# ============================
# Task 4: ChatPromptTemplate System
# ============================

def get_chat_prompt(role, topic):

    if role == "teacher":
        system_msg = "You are a helpful teacher."
    elif role == "interviewer":
        system_msg = "You are an interviewer asking questions."
    elif role == "motivator":
        system_msg = "You are a motivational speaker."
    else:
        system_msg = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("human", "Talk about {topic}")
    ])

    return chat_template.format_messages(topic=topic)


print("Task 4 Output:")
msgs = get_chat_prompt("teacher", "Neural Networks")
for m in msgs:
    print(m)
print("\n")


# ============================
# Task 5: Input Validation Layer
# ============================

def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience → default: beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone → default: formal")
        tone = "formal"

    return audience, tone


print("Task 5 Output:")
print(validate_inputs("kids", "cool"))
print("\n")


# ============================
# Task 6: Prompt Generator App
# ============================

def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Generate 3 interview questions about {topic} for {audience} in a {tone} tone"
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )

    else:
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} tone"
        )

    return template.format(topic=topic, audience=audience, tone=tone)


print("Task 6 Output:")
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
print("\n")


# ============================
# Task 7: Template Reusability Test
# ============================

reuse_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

print("Task 7 Outputs:")
print(reuse_template.format(topic="AI", audience="beginner", tone="formal"))
print(reuse_template.format(topic="Blockchain", audience="expert", tone="formal"))
print(reuse_template.format(topic="Python", audience="intermediate", tone="casual"))
print(reuse_template.format(topic="Cybersecurity", audience="beginner", tone="fun"))
print(reuse_template.format(topic="Data Science", audience="expert", tone="casual"))
```
